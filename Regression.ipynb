```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/StudentsPerformance.csv')
```

```python
df.isnull().sum()
```

```python
df["math score"].mean()
```

```python
df["math score"].median()
```

```python
df["math score"].std()
```

```python
plt.hist(df["math score"])
plt.title("Histogram of Math Scores")
plt.xlabel("Math Score")
plt.show()
```

```python
plt.scatter(df["reading score"],df["math score"])
plt.title("Scatterplot of Math Scores")
plt.xlabel("reading score/study hours")
plt.ylabel("Math Score")
plt.show()
```

```python
x = df[["reading score"]]
y = df["math score"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
predicted = model.predict(x)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["reading score"], df["math score"])
plt.plot(df["reading score"], predicted,color="r")
plt.title("Simple Linear Regression")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
