```python
def calculate_total(marks):
    return sum(marks)
```

```python
def calculate_percentage(total, num_subjects):
    return total / num_subjects
```

```python
def grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "F"
```

```python
marks=[66,78,89,33,92]
total=calculate_total(marks)
print("total:",total)
percentage=calculate_percentage(total, 5)
print("percentage:",percentage,'%')
grades=("Grades:",grade(percentage))
print(grades)
```

```python
count=0
with open(path+"/sal.txt","r") as salfile:
  s=salfile.readlines()
  for i in s:
    sort=i.strip().split(",")

    # print(sort)

    if int(sort[1]) > 50000:
      print(sort)
      count +=1

print ("no. of employees with greater salary",count)
```

```python
p=0
a=0
with open(path+"/attendence.txt","r") as ab:
  cont=ab.readlines()
  print("Absent Student")
  for i in cont:
    d=i.strip().split(",")
    # print(d)
    stu=d[0]
    status=d[1].strip()
    # print(status)

    if status == 'absent':
      print(stu+status)
      a += 1
    else:
      p += 1

print("no. of absents:",a)
print("no. of present:",p)




```
